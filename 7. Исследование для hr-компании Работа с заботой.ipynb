## Описание проекта

Заказчик проекта — компания «Работа с заботой», основная цель - помочь бизнесу оптимизировать управление персоналом. Для этого будет использовано машинное обучение, чтобы быстрее и точнее отвечать на вопросы бизнеса.

Компания предоставила данные о своих сотрудниках, включая уровень их удовлетворенности работой. Эта информация была получена из обратной связи, где сотрудники заполняют опросник, и на основе его результатов рассчитывается уровень их удовлетворенности от 0 до 1.

Первая задача заключается в построении модели, которая сможет предсказать уровень удовлетворенности сотрудника на основе предоставленных данных. Это важно для бизнеса, поскольку удовлетворенность работой напрямую влияет на отток сотрудников. Предсказание оттока является одной из важнейших задач HR-аналитиков, поскольку внезапные увольнения могут представлять риски для компании, особенно если уходит важный сотрудник.

Вторая задача заключается в построении модели, которая сможет предсказать, уволится ли сотрудник из компании на основе предоставленных данных. Это также важно для минимизации рисков, связанных с потерей ключевых сотрудников.

**Описание данных**

Данные, представленные заказчиком, включают следующие признаки:

- `id` — уникальный идентификатор сотрудника
- `dept` — отдел, в котором работает сотрудник
- `level` — уровень занимаемой должности
- `workload` — уровень загруженности сотрудника
- `employment_years` — длительность работы в компании (в годах)
- `last_year_promo` — информация о том, было ли повышение за последний год
- `last_year_violations` — информация о том, нарушал ли сотрудник трудовой договор за последний год
- `supervisor_evaluation` — оценка качества работы сотрудника, которую дал руководитель
- `salary` — ежемесячная зарплата сотрудника
- `job_satisfaction_rate` — уровень удовлетворённости сотрудника работой в компании, целевой признак

**Так как работа состоит из двух задач, создадим несколько собственных функций, чтобы автоматизировать аналогичные действия.**

Функция для вывода основной информации о датафрейме:

````python
def information_data(data):
    display('Первые десять строк датафрейма:')
    display(data.head(10))
    display('Общая информация о датафрейме:')
    display(data.info())
    display('Описание данных:')
    display(data.describe())
    display('Количество пропусков:')
    display(data.isna().sum())
    display('Количество дубликатов:')
    display(data.duplicated().sum())
````

Функция для постоения гистограммы и диаграммы размаха для проведения исследовательского анализа количественных признаков:

````python
def hist_boxplot(data, features, column_names):
    for col in features:
        fig, axs = plt.subplots(1, 2, figsize=(20, 5))

        # Гистограмма
        axs[0].hist(data[col], color='lightseagreen', edgecolor='black', alpha=0.5)
        axs[0].set_title(f'Гистограмма для {column_names[col]}', fontsize=15)
        axs[0].set_xlabel(column_names[col])  
        axs[0].set_ylabel('count')  

        # Диаграмма размаха
        sns.boxplot(x=data[col], ax=axs[1], color='lightseagreen')
        axs[1].set_title(f'Диаграмма размаха для {column_names[col]}', fontsize=15)
        axs[0].set_xlabel(column_names[col])  

        plt.show();
````

Функция для построения круговой диаграммы для категориальных признаков:

````python
colors = ['lightpink', 'powderblue', 'wheat', 'lightcoral', 'palegreen', 'skyblue']

def plot_pie(data, features, column_names, ylbl=None, xlbl=None):
    for col in features:
        plt.figure(figsize=(10, 7))
        data[col].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=colors)
        plt.title(f'Круговая диаграмма {column_names[col]}')
        plt.xlabel(xlbl)
        plt.ylabel(ylbl)
        plt.show();
````

Функция для создания наложенных столбчатых диаграмм для тренировочной и тестовой выборки:

````python
def overlay_barplot(train_data, test_data, features, feature_name_ru):
    for col in features:
        fig, ax = plt.subplots(figsize=(10, 5))

        # Столбчатая диаграмма для тренировочных данных
        sns.countplot(y=col, data=train_data, color='lightseagreen', alpha=0.5, label='Train', ax=ax)

        # Столбчатая диаграмма для тестовых данных
        sns.countplot(y=col, data=test_data, color='seagreen', alpha=0.5, label='Test', ax=ax)

        ax.set_title(f'Распределение признака {feature_name_ru[col]} в тренировочных и тестовых данных', fontsize=15)
        ax.legend(loc='lower right')

        plt.show();
````

- **Обобщение:** Модели, которые хорошо работают на данных с похожим распределением, обычно лучше обобщаются на новые данные. Это означает, что они могут быть более эффективными при прогнозировании результатов на данных, которые они ранее не видели.

### Шаг 4. Подготовка данных

Выполним подготовку признаков в пайплайне, дополнив его шагом предобработки. При кодировании учтем особенности признаков и моделей и используем как минимум два кодировщика.

Перечислим особенности данных:

- Три признака:
`dept`, `last_year_promo`, `last_year_violations` — нужно кодировать с помощью `OneHotEncoder`.
- Два признака:
`level`, `workload` — нужно кодировать с помощью `OrdinalEncoder`.
- 3 количественных признака:
`employment_years`, `salary`, `supervisor_evaluation` — нужно масштабировать.
В признаках встречаются пропуски, обработаем их в пайплайне.
- Целевой признак — `job_satisfaction_rate` масштабировать или кодировать не будем: в нашем случае, где целевой признак — это непрерывная переменная в диапазоне от 0 до 1, масштабирование не требуется.

Создадим первый пайплайн для подготовки данных. Сначала введем обозначения для типов исходных данных:

````python
ohe_columns = [
    'dept',
    'last_year_promo',
    'last_year_violations'
]
ord_columns = [
    'level',
    'workload'
]
````

````python
num_columns = ['employment_years', 'supervisor_evaluation', 'salary']
````

Пайплайн для подготовки данных:

````python
# SimpleImputer + OHE
ohe_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ohe',
            OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
        )
    ]
)
# SimpleImputer + ORD + SimpleImputer

'''
При порядковом кодировании необходимо задать порядок категоризации.
Поэтому в параметре categories последовательно перечислим
уровни категорий для каждого признака из списка ord_columns.
'''

ord_pipe = Pipeline(
    [
        (
            'simpleImputer_before_ord',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',
            OrdinalEncoder(
                categories=[
                    ['junior', 'middle', 'senior'],
                    ['low', 'medium', 'high']
                ],
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
        (
            'simpleImputer_after_ord',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)
# Объединяем все шаги подготовки данных
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', 'passthrough', num_columns)
    ],
    remainder='passthrough'
)
````

Разработали пайплайн для подготовки данных. Он включает в себя обработку пропущенных значений, кодирование категориальных переменных и масштабирование числовых переменных.

Категориальные признаки:
- `dept`, `last_year_promo`, `last_year_violations` будут кодироваться с помощью `OneHotEncoder`.

Категориальные признаки, в которых порядок имеет значение (больше-меньше, лучше-хуже)
- `level`, `workload` будут обрабатываться с помощью `OrdinalEncoder`.

Количественные признаки:
`employment_years`, `supervisor_evaluation`, `salary` будем масштабировать, два масштабировщика `StandardScaler` и `MinMaxScaler` зададим в пайплайне на этапе обучения моделей.

Целевой признак — `job_satisfaction_rate` оставили без преобразования.

### Шаг 5. Обучение моделей

Обучим модели: линейные модели `Ridge`, `Lasso` и модель случайного леса `RandomForestRegressor`. Подберем гиперпараметры для как минимум одной модели и выберем лучшую, проверим её качество на основе метрики `SMAPE` (симметричное среднее абсолютное процентное отклонение), которая вычисляется по формуле:

$$
SMAPE = \frac{100}{n} \sum_{i=1}^{n} \frac{|y_i - \hat{y}_i|}{(|y_i| + |\hat{y}_i|)/2}
$$

где:
- $y_i$ — фактическое значение целевого признака для объекта с порядковым номером $i$ в выборке;
- $\hat{y}_i$ — предсказанное значение целевого признака для объекта с порядковым номером $i$ в выборке;
- $n$ — количество объектов в выборке;
- $\sum_{i=1}^{n}$ — сумма значений, полученная в результате операций, которые следуют за этим знаком, для всех объектов с порядковым номером от $i$ до $n$ в выборке.

Цель — достичь значения `SMAPE` ≤ 15 на тестовой выборке.

Обучать модель будем на наборе данных `train_job_satisfaction_rate`, предварительно разделив на входные признаки и целевой `job_satisfaction_rate`. Тестировать будем на тестовых данных, представленных заказчиком: `test_features` (содержит входные признаки. Ранее объединили с целевым признаком `test_target_job_satisfaction_rate`). Перед началом моделирования удалим дубликаты из тренировочной выборки: они могут привести к переобучению модели, поскольку искажают распределение данных. Это позволяет убедиться, что модель обучается на уникальных данных и может обобщать, а не просто запоминать данные.

````python
# Проверка на дубликаты
print('Количество дубликатов в тренировочной выборке:', train_job_satisfaction_rate.duplicated().sum())

# Удаление дубликатов
train_job_satisfaction_rate = train_job_satisfaction_rate.drop_duplicates()

# Проверка, что дубликаты были удалены
print('Количество дубликатов в тренировочной выборке после удаления:', train_job_satisfaction_rate.duplicated().sum())
````

````python
# Разделяем данные на обучающую и тестовую выборки
X_train = train_job_satisfaction_rate.drop('job_satisfaction_rate', axis=1)
y_train = train_job_satisfaction_rate['job_satisfaction_rate']
X_test = test_features.drop('job_satisfaction_rate', axis=1)
y_test = test_features['job_satisfaction_rate']
````

````python
# Создаем итоговый пайплайн
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', RandomForestRegressor(random_state=RANDOM_STATE))
])
````

````python
# Задаем сетку параметров для поиска
param_grid = [
    {
        'models': [RandomForestRegressor(random_state=RANDOM_STATE)],
        'models__n_estimators': [100, 200, 300],  # количество деревьев
        'models__max_depth': range(5, 10),  # максимальная глубина деревьев
        'models__max_features': range(5, 10),  # максимальное количество признаков для разделения
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    {
        'models': [Ridge(random_state=RANDOM_STATE)],
        'models__alpha': np.logspace(-4, 4, 20),  # параметр регуляризации
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    {
        'models': [Lasso(random_state=RANDOM_STATE)],
        'models__alpha': np.logspace(-4, 4, 20),  # параметр регуляризации
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]
````

````python
# создаем функцию для рассчета метрики SMAPE
def smape(y_true, y_pred):
    epsilon = 1e-6  # небольшая константа, чтобы избежать поялвение nan
    return 100/len(y_true) * np.sum(2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred) + epsilon))

smape_scorer = make_scorer(score_func=smape, greater_is_better=False)
````

````python
# Создаем экземпляр GridSearchCV для поиска лучших параметров
grid_search = GridSearchCV(
    pipe_final,
    param_grid,
    cv=5,
    scoring=smape_scorer,
    n_jobs=-1
)
grid_search.fit(X_train, y_train)
grid_search.best_params_
````

**Сохраняем лучшую модель**

````python
best_model = grid_search.best_estimator_

# Получаем модель из пайплайна
model = best_model.named_steps['models']
````

**Выведем значение метрики SMAPE для тренировочной и тестовой выборки.**

````python
print(f'Метрика SMAPE лучшей модели при кросс-валидации на обучающих данных: {round(grid_search.best_score_*(-1),2)}')
````

````python
predictions = best_model.predict(X_test)

print(f'Метрика SMAPE лучшей модели на тестовой выборке: {round(smape(y_test, predictions),2)}')

# Вычисление ошибки модели
model_error = abs(smape_scorer(best_model, X_test, y_test))

print('Ошибка лучшей модели:', round(model_error,2))
````

Проведем проверку нашей модели на адекватность, сравнивая её предсказательные способности с константной моделью. Константная модель, в данном случае `DummyRegressor`, это базовая модель, которая не использует никакой информации о данных для предсказания. Она просто возвращает среднее значение целевой переменной из обучающего набора данных для всех предсказаний.

Это важный шаг, поскольку он позволяет установить минимальный порог эффективности для нашей модели. Если наша модель не сможет превзойти этот базовый уровень, это может указывать на то, что она не достаточно хорошо обучилась на данных и, возможно, требует дальнейшей оптимизации или пересмотра подхода.

Для оценки адекватности используем ту же метрику — `SMAPE`.

````python
# Создание константной модели
dummy = DummyRegressor(strategy='mean')

# Обучение константной модели
dummy.fit(X_train, y_train)

# Вычисление ошибки модели
dummy_error = abs(smape_scorer(dummy, X_test, y_test))

print('Ошибка константной модели:', round(dummy_error, 2))
````


Сравним ошибку константной модели с ошибкой нашей лучшей модели:


````python
if model_error < dummy_error:
    print('Точность лучшей модели выше, чем точность константной модели.')
elif model_error == dummy_error:
    print('Точность лучшей модели такая же, как точность константной модели.')
else:
    print('Точность константной модели выше, чем точность лучшей модели.')
````

### Шаг 6. Обучение модели

Обучим как минимум три модели и подберем гиперпараметры для них. Проверим качество лучшей модели.

Метрика оценки качества в этой задаче — `ROC-AUC`. Цель — достичь значения `ROC-AUC` ≥ 0.91 на тестовой выборке.

````python
# Разделяем данные на обучающую и тестовую выборки
X_train = train_quit.drop('quit', axis=1)
y_train = train_quit['quit']
X_test = test_features.drop('quit', axis=1)
y_test = test_features['quit']
````

````python
# Создаем экземпляр LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
````

````python
# Создаем итоговый пайплайн
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor), 
    ('models', RandomForestClassifier(random_state=RANDOM_STATE)) 
])
````

````python
# Задаем сетку параметров для поиска
param_grid = [
    {
        'models': [RandomForestClassifier(random_state=RANDOM_STATE)],
        'models__n_estimators': [100, 150, 200],  # количество деревьев
        'models__max_depth': range(5, 8),  # максимальная глубина деревьев
        'models__max_features': range(5, 8),  # максимальное количество признаков для разделения
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    {
        'models': [SVC(random_state=RANDOM_STATE, probability=True)], 
        'models__kernel': ['linear', 'sigmoid', 'poly', 'rbf'], 
        'models__degree': range(2, 4), 
        'models__C': np.linspace(0.1, 1, 10), 
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE, solver='liblinear', penalty='l1')], 
        'models__C': range(1, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough'] 
    },
]
````

````python
# Создаем экземпляр RandomizedSearchCV для поиска лучших параметров
randomized_search = RandomizedSearchCV(
    pipe_final,  
    param_grid,  
    cv=5, 
    scoring='roc_auc', 
    n_jobs=-1
)
````

````python
randomized_search.fit(X_train, y_train) 
randomized_search.best_params_
````

````python
# Выводим лучшую модель и её параметры
print('Лучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
print('Метрика лучшей модели на валидационной выборке:', round(randomized_search.best_score_, 2))
````

````python
# Оцениваем модель на тестовом наборе
y_test_pred_proba = randomized_search.predict_proba(X_test)[:, 1] 
print(f'Метрика ROC-AUC на тестовой выборке: {round(roc_auc_score(y_test, y_test_pred_proba), 2)}')
````

Проверим модель на адекватность с помощью константной модели:

````python
y_test_pred = randomized_search.predict(X_test)

model_error = roc_auc_score(y_test, y_test_pred)

print('Точность лучшей модели:', round(model_error, 2))
````
